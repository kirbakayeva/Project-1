## Home Sales Analysis in different cities

* The module analyzes the Home sales with in different cities with relation to the crime rate

## import dependencies

```python
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
import os
import random
from pprint import pprint
```

## Read the dataset - Crime and Housing Data

```python
# read the dataset - Crime and Housing Data
file_path = os.path.join("..", "Resources", "Redfin_CrimeDetailedset.csv")
housingCrimeDF = pd.read_csv(file_path)
print(f"Data Shape: {housingCrimeDF.shape}")
housingCrimeDF.head()
```

## build the dataframe for analysis

```python
# build the data frames for analysis

houstondf = housingCrimeDF[housingCrimeDF['City'] == 'Houston']
friendswooddf = housingCrimeDF[housingCrimeDF['City'] == 'Friendswood']
leagueCitydf = housingCrimeDF[housingCrimeDF['City'] == 'League City']
sugarlanddf = housingCrimeDF[housingCrimeDF['City'] == 'Sugar Land']
baytowndf = housingCrimeDF[housingCrimeDF['City'] == 'Baytown']
```

## Plot the regression graph

```python
# Perform a linear regression on crime vs units sold
(slopeH, interceptH, rvalueH, pvalueH, stderrH) = stats.linregress(houstondf['Total Crime Per Capita [10,000]'], houstondf['Homes Sold'] )
# Get regression values
regress_valuesHouston = houstondf['Total Crime Per Capita [10,000]'] * slopeH + interceptH

 # Create line equation string
line_eqH = "Houston : y = " + str(round(slopeH,2)) + "x +" + str(round(interceptH,2))

# Perform a linear regression on crime  vs units sold
(slopeF, interceptF, rvalueF, pvalueF, stderrF) = stats.linregress(friendswooddf['Total Crime Per Capita [10,000]'], friendswooddf['Homes Sold'])
# Get regression values
regress_valuesFriendswood = friendswooddf['Total Crime Per Capita [10,000]'] * slopeF + interceptF
 # Create line equation string
line_eqF = "Friendswood : y = " + str(round(slopeF,2)) + "x +" + str(round(interceptF,2))


# Perform a linear regression on crime  vs units sold
(slopeL, interceptL, rvalueL, pvalueL, stderrL) = stats.linregress(leagueCitydf['Total Crime Per Capita [10,000]'], leagueCitydf['Homes Sold'])
# Get regression values
regress_valuesLeagueCity = leagueCitydf['Total Crime Per Capita [10,000]'] * slopeL + interceptL
 # Create line equation string
line_eqL = "League City : y = " + str(round(slopeL,2)) + "x +" + str(round(interceptL,2))


# Create Plot
ax1 = houstondf.plot(kind='scatter', x='Total Crime Per Capita [10,000]', y='Homes Sold', color ='skyblue',  alpha=0.5, figsize=(10, 7))
friendswooddf.plot(kind='scatter', x='Total Crime Per Capita [10,000]', y='Homes Sold', color ='grey', alpha=0.5, figsize=(10, 7),  ax=ax1)
leagueCitydf.plot(kind='scatter', x='Total Crime Per Capita [10,000]', y='Homes Sold', color ='brown', alpha=0.5, figsize=(10, 7),  ax=ax1)

#plot regression lines 
plt.plot(houstondf['Total Crime Per Capita [10,000]'],regress_valuesHouston,"b-")
plt.plot(friendswooddf['Total Crime Per Capita [10,000]'],regress_valuesFriendswood,"k-")
plt.plot(leagueCitydf['Total Crime Per Capita [10,000]'],regress_valuesLeagueCity,"r-")

#plot regression text
plt.text(400,250,line_eqH, fontsize=10,color="green")
plt.text(400,235,line_eqF, fontsize=10,color="green")
plt.text(400,220,line_eqL, fontsize=10,color="green")


# legend, title and labels.
plt.legend(labels=['Houston Regression Line', 'Friendswood Regression Line', 'League City Regression Line', 'Houston', 'Friendswood', 'League City'])
#plt.legend(labels=['Houston Regression Line', 'Sugar Land Regression Line', 'Houston', 'Sugar Land'])
plt.title('Home Sales vs Crime', size=14)
plt.xlabel('Total Crime Per Capita [10,000]')
plt.ylabel('House Units Sold Per Capita [10,000]');

#set xlimits
plt.xlim(0,600)

# Save the plotted figures as .pngs
output_path = os.path.join("..", "Images", "Crime and Units Sold for Houston Friendswood And LeagueCity.png")
plt.savefig(output_path)

#show the graph
plt.show()

#print p values
print(f"The pValue for Houston is: {round(pvalueH,4)}")
print(f"The pValue for Friendswood  is: {round(pvalueF,4)}")
print(f"The pValue for League City  is: {round(pvalueL,4)}")
```
